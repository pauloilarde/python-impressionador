## Funcionamento

- Vamos ler um arquivo real, com a Base de Dados de Vendas da Empresa "Contoso"

```python
import pandas as pd

# o nome do arquivo quando baixa o csv está com espaço
# para funcionar é preciso remover o espaço extra
# vendas_df = pd.read_csv(r'Contoso - Vendas  - 2017.csv', sep=';')
vendas_df = pd.read_csv(r'Contoso - Vendas - 2017.csv', sep=';')

# exibe o dataframe inteiro
vendas_df

# exibe apenas uma coluna do dataframe
vendas_df['ID Cliente']

# falando sobre linhas
# vendas_df[0] # -> NÃO FUNCIONA ASSIM PARA DATAFRAMES, dá erro: KeyError: 0
# vendas_df[:0] #-> isso funcionaria, pegue "até" o ínedice zero, ou seja: o cabeçalho
# vendas_df[:3] #-> pega até a linha de índice 3 do dataframe, incluindo o cabeçalho

# falando sobre colunas
# criando novo dataframe com três colunas
# como se fosse um filtro no dataframe original
vendas_df[['Numero da Venda', 'Data da Venda', 'ID Produto']]

# obs: aqui estamos trabalhando com o dataframe pequeno
# até que tem bastante linhas, mas tem poucas colunas
# por isso quando pedimos para exibir a tabela, ele exibe bonitinha
# se estivessemos trabalhando com um dataframe maior ele exibe em outro formato


# e aí, para pegar uma linha de uma determinada coluna
vendas_df['ID Produto'][0]
# no vídeo, mostrou o número 981
# no meu caso, mostrou 
# np.int64(981)

# falando sobre o comantário um pocuo acima:
# ao se deparar com uma base de dados pela primeira vez
# temos que entender o que tem nela
# exibindo a tabela é um jeito de fazer isso, como fizemos acima
# basta escrever o nome do dataframe, vendas_df
# mas não fica um bom visual em tabelas muito grandes
# na célula abaixo, usaremos o método .info()
```
