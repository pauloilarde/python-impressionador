## Arrays

Um array é uma estrutura de dados que armazena valores do mesmo tipo. Em Python, isso é uma grande vantagem porque economiza espaço e permite operações mais eficientes. Vamos criar um array.



````python
import numpy as np

# Criação de um array
# array = np.array([1, 2, 3, 4, 5])
# mudando array de números para letras, por didática
# para não confundir o elemento com índice
array = np.array(["a", "b", "c", "d", "e"])
print(array)
# ['a' 'b' 'c' 'd' 'e']
````

## Conhecendo um array: indexação e slice





````python
print(array[0])
# a
# primeiro elemento do array
# o índice funciona começando do zero 
# e a contagem pode ser feita a partir do último elemento, 
# o último elemento é índice -1
````

````python
print(array[1:4])
# ['b' 'c' 'd']
# um recorte do segundo elemento (pois a contagem começa em zero)
# até o quarto elemento
# note que fomos até o índice 4, que seria o quinto elemento
# mas ele exclui o valor final do recorte
````

````python
print(array[0:-1])
# ['a' 'b' 'c' 'd']

# um recorte do primeiro ao último elemento 
# mas ele exclui o valor final do recorte
````

````python
print(array[0:-1:2])
# ['a' 'c']

# um recorte do primeiro ao último
# aqui estamos avançando de dois em dois elementos
# chamamos esse avanço de "passo"
# o padrão é avançar em passo um - de um em um
# se o array fosse maior, seria mais perceptível 
````

````python
print(array[0::2])
# ['a' 'c' 'e']

# um recorte do primeiro ao último
# mas aqui está omitindo o fim, então ele inclui o último elemento
# e avançando de dois em dois
````

````python
print(array[::2])
# ['a' 'c' 'e']

# do mesmo jeito que omitimos o fim
# também podemos omitir o começo
# nos anteriores nós começamos no elemento zero
# mas poderíamos ter começado de qq elemento do array
# se omitir o início, ele começa do zero
# aqui também avançamos de dois em dois
````

````python
print(array[::])
# ['a' 'b' 'c' 'd' 'e']

# esse caso aqui traz o array todo
# omitimos o começo, omitimos o fim e omitimos o passo
````

````python
print(array[::-1])

# ['e' 'd' 'c' 'b' 'a']

# também pegou o array todo
# omitimos o começo e o fim
# e o passo é -1, então ele vai de trás pra frente
````
