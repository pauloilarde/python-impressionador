```python
import numpy as np

# Pesos de pacotes (em kg) - tolerância: 1 desvio padrão
pesos = np.array([10.2, 10.5, 10.3, 9.8, 10.1, 11.5, 10.0, 9.5, 10.4])

# Encontre pacotes fora da tolerância (> média + 1σ OU < média - 1σ)
media = np.mean(pesos)
print(f"Média: {media:.2f}")

desvio_padrao = np.std(pesos)
print(f"Desvio padrão: {desvio_padrao:.2f}")

fora_tolerancia = np.where((pesos > (media + desvio_padrao)) | (pesos < (media - desvio_padrao)))
print(pesos[fora_tolerancia])

# Média: 10.26
# Desvio padrão: 0.53
# (array([], dtype=int64),)
```

```python
# Temperaturas diárias (°C) - alerta se > 30°C OU < 15°C
temperaturas = np.array([22, 25, 31, 18, 28, 14, 26, 33, 20])


# Encontre dias com alerta
temp_alerta = np.where((temperaturas > 30) | (temperaturas < 15))
print(f"Temperaturas alerta: {temperaturas[temp_alerta]}")
```

```python
# Vendas diárias - identifique outliers (> média + 1.5σ OU < média - 1.5σ)
vendas = np.array([150, 160, 155, 148, 152, 200, 158, 100, 151])

media = np.mean(vendas)
print(f"Média vendas: {media:.2f}")

desvio_padrao = np.std(vendas)
print(f"Desvio padrão: {desvio_padrao:.2f}")
# Encontre os valores outliers

outliers = np.where((vendas > media + desvio_padrao) | (vendas < media - desvio_padrao))
print(f"Dias outlier: {vendas[outliers]}")
```
